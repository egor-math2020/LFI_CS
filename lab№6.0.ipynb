```python
#упражнение 7
B=pd.DataFrame(f.groupby(["State"]).agg({"Total day calls": "mean" , "Total eve calls": "mean"}))
B
```

```python
# упражнение 8
import numpy as np
w = np.array(B["Total day calls"] - B["Total eve calls"]) 
r =[]
for i in range(len(w)):
  if w[i] >= 0:
    w == True
    r.append("True")
  else: 
    w == False
    r.append("False")
B["comparison"] = r
B
```
